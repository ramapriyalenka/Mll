```python
import pandas as pd
df=pd.DataFrame(
    {
    'Height':[7,5,6,60,100,200,5,150],
    'Name':["Human","Human","Human","Tree","Tree","Tree","Human","Tree"]
    })
df
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
x=df[['Height']].values
x
```

```python
kmeans.fit(x)
kmeans.predict(x)
```

```python
kmeans.inertia_
```

```python
inr=[]
for k in [1,3,5,8]:
    kmeans = KMeans(n_clusters=k,random_state=42)
    kmeans.fit(x)
    inr.append(kmeans.inertia_)
print(inr)
```

```python
%matplotlib notebook
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.plot([1,3,5,8],inr)
plt.show()
```
