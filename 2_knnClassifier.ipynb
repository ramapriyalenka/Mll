```python
df=pd.DataFrame(c_data['data'])
df.columns=c_data['feature_names']
df['Target']=c_data['target']
df.head(10)
```

```python
x=df[['mean area','mean perimeter','mean concavity','mean symmetry']]
y=df[['Target']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
x_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knnModel=KNeighborsClassifier(n_neighbors=2)
knnModel.fit(x_train,y_train)
print("Training Score")
print('\t',knnModel.score(x_train,y_train)*100)
print("Testing Score")
print('\t',knnModel.score(x_test,y_test)*100)
```

```python
scores={}
for neighbor in range(1,21):
    knnModel=KNeighborsClassifier(n_neighbors=neighbor)
    knnModel.fit(x_train,y_train)
    scores[neighbor]=knnModel.score(x_test,y_test)*100
scores
    
```
