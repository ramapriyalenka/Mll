```python
import pandas as pd
%matplotlib notebook
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
```

```python
df=pd.DataFrame(cancer['data'])
df.columns=cancer['feature_names']
df['Target']=cancer['target']
df.head(10)
```

```python
from sklearn.decomposition import PCA
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sclr=MinMaxScaler()
```

```python
X_data=df[df.columns[:-1]].as_matrix()
X_data
```

```python
X_scaled = sclr.fit_transform(X_data)
X_scaled
```

```python
pcaModel=PCA(n_components=3)
#n_component means no of princioalcomponents or features
X_pca=pcaModel.fit_transform(X_scaled)
```

```python
len(X_pca)
```

```python
X_pca_df=pd.DataFrame(X_pca).head()
X_pca_df.head()
```

```python
pcaModel.explained_variance_ratio_
```

```python
import matplotlib.pyplot as plt
```

```python
X_pca_df["Target"]=df["Target"]
X_pca_df
```

```python
group0 = X_pca_df[X_pca_df["Target"]==0]
group0
```

```python
group1 = X_pca_df[X_pca_df["Target"]==1]
group1.head()
```

```python
plt.figure()
plt.scatter(
    group0[group0.columns[0]],
    group0[group0.columns[1]],
    marker="*",c="r",label="group0")
plt.scatter(
    group1[group1.columns[0]],
    group1[group1.columns[1]],
    marker="*",c="b",label="group1")
plt.legend()
plt.show()
```
