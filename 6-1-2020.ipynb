```python
def sleep_in(weekday, vacation):
    
    if not weekday or vacation:
        return True
    else:
        return False
sleep_in(1,0)
```

```python
def monkey_trouble(a_smile, b_smile):
    if a_smile and b_smile:
        return True
    if not a_smile and not b_smile:
        return True
    return False
monkey_trouble(1,1)
```
