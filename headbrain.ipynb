```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("headbrain.csv")
data
```

```python
# collecting X and Y
X=data['Head Size(cm^3)'].values
Y=data['Brain Weight(grams)'].values

Y

```

```python
# Mean X and Y
mean_x = np.mean(X)
print(mean_x)
mean_y = np.mean(Y)
print(mean_y)
```

```python
n=len(X)
n
```

```python
num=0
denom=0
for i in range(n):
    num+=(X[i]-mean_x)*(Y[i]-mean_y)
    denom+=(X[i]-mean_x)**2
b1 = num/denom
b0 = mean_y-(b1*mean_x)
print(b1)
print(b0)
```

```python
max_x=np.max(X)+100
max_x
min_x=np.max(Y)-100
min_x
```

```python
x=np.linspace(min_x,max_x,1000)
y=b0+b1*x
#plotting line
plt.plot(x,y,color="Black",label="Regression Line")
plt.scatter(X,Y,c="Blue",label="Scatter Plot")
plt.xlabel('Head Size in cm3')
plt.ylabel('Brain Weight in grams')
plt.legend()
plt.show()
```

```python
ss_tot=0
ss_res=0
for i in range(n):
    y_pred=b0+b1*x[i]
    ss_tot+=(y[i]-mean_y)**2
    ss_res+=(y[i]-y_pred)**2
r2=1-(ss_res/ss_tot)
print("R2 Score")
print(r2)
```
