```python
import pandas as pd
```

```python
horror = [3,4,5, 2,2,3, 4,5,0, 1,0,1]
family = [1,2,3, 4,4,5, 0,1,3, 4,0,1]
```

```python
df = pd.DataFrame({'horror':horror,'family':family})
```

```python
df.head()
```

```python
df
```

```python
df.shape
```

```python
from sklearn.cluster import KMeans
```

```python
%matplotlib notebook
```

```python
import matplotlib.pyplot as plt
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(df.values)
```

```python
pred = kmeans.predict(df.values)
df['pred'] = pred
```

```python
df
```

```python
group0 = df[df['pred']==0]
group0.head()
```

```python
group0.shape
```

```python
group1 = df[df['pred']==1]
group1.head()
```

```python
group1.shape
```

```python
group2 = df[df['pred']==2]
group2.head()
```

```python
group3 = df[df['pred']==3]
group3.shape
```

```python
plt.figure()
plt.scatter(
    group0.horror.values,
    group0.family.values,
    c='r',
    label='group0'
)
plt.xlim(-7,17)
plt.ylim(-7,17)
plt.scatter(group2.horror.values,group2.family.values,c='b',label='group2')
plt.scatter(group3.horror.values,group3.family.values,c='yellow',label='group3')
plt.scatter(group1.horror.values,group1.family.values,c='g',label='group1')

plt.xlabel('horror')
plt.ylabel('family')
plt.legend()
plt.show()
```
