```python
import pandas as pd
l=[1,2,8,9,11]
l1=[100,101,90,92,200]
data=pd.DataFrame({'first':l,'second':l1})
data
```

```python
from sklearn.preprocessing import normalize
mydata=normalize(data)
data2=pd.DataFrame(mydata,columns=['first','second'])
data2
```

```python
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt
plt.title("hierarchy")
d=shc.dendrogram(shc.linkage(data2,method='average'))
```

```python
import pandas as pd
df=pd.read_csv("Wholesale customers data.csv")
df
```
