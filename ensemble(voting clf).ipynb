```python
import pandas as pd
```

```python
df=pd.read_table("fruit_data_with_colors.txt")
df.head()
```

```python
df.loc[df["fruit_label"]!=1,"fruit_label"]=0
df#.head()
```

```python
df['fruit_label'].value_counts()
```

```python
x=df[['mass','width','height']]
x.head()
```

```python
y=df['fruit_label']
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import VotingClassifier
```

```python
ir_cls=LogisticRegression()
rf_cls=RandomForestClassifier()
nb_cls=GaussianNB()
```

```python
vclf=VotingClassifier(estimators=[('ir',ir_cls),('tf',rf_cls),('ga',nb_cls)],n_jobs=-1)
```

```python
from sklearn.metrics import accuracy_score
for cls in [ir_cls,rf_cls,nb_cls,vclf]:
    cls.fit(x_train,y_train)
    y_pred=cls.predict(x_test)
    print("accuracy score",cls.__class__.__name__,accuracy_score(y_test,y_pred))
```

## Zoo data set

```python
df=pd.read_csv("zoo.csv")
df
```

```python
x=df[['aquatic','predator','toothed','backbone']]
x
```

```python
y=df['class_type']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import VotingClassifier
```

```python
dtc=DecisionTreeClassifier()
svm=LinearSVC()
knn=KNeighborsClassifier()
```

```python
vclf=VotingClassifier(estimators=[('ir',dtc),('tf',svm),('ga',knn)],n_jobs=-1,voting='hard')
```

```python
from sklearn.metrics import accuracy_score
for cls in [dtc,svm,knn,vclf]:
    cls.fit(x_train,y_train)
    y_pred=cls.predict(x_test)
    print("accuracy score",cls.__class__.__name__,accuracy_score(y_test,y_pred))
```
