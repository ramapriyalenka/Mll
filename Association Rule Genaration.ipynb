```python
import pandas as pd
```

```python
df=pd.DataFrame({
    'Transaction ID':[1,2,3,4,5],
    'Bread':[1,1,0,1,1],
    'Milk':[1,0,1,1,1],
    'Beer':[0,1,1,1,0],
    'Eggs':[1,0,1,0,0],
    'cola':[0,0,1,0,0]
})
df
```

```python
from mlxtend.frequent_patterns import (
     apriori,
    association_rules
)
```

```python
data_df = df[df.columns[1:]]
```

```python
freq_items = apriori(data_df,use_colnames=True)
freq_items
```

                                    
support for item x                -------------------------
                                       total transaction

```python
rules = association_rules(freq_items,min_threshold=0.2)
rules
```
