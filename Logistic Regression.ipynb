```python
import pandas as pd
```

```python
df=pd.read_table("fruit_data_with_colors.txt")
df.head()
```

```python
df.loc[df["fruit_label"]!=1,"fruit_label"]=0
df#.head()
```

```python
df['fruit_label'].value_counts()
```

```python
x=df[['mass','width','height']]
x.head()
```

```python
y=df['fruit_label']
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg=LogisticRegression()
logreg.fit(x_train,y_train)
```

```python
logreg.score(x_train,y_train)
```

```python
logreg.score(x_test,y_test)
```

```python
x_test.head()
```

```python
logreg.predict([[80,5.8,4.3],[186,7.2,9.2],[84,6.0,4.6],[7.1,7.5,0.79]])
```

```python
logreg.predict_proba([[80,5.8,4.3],[186,7.2,9.2],[84,6.0,4.6],[7.1,7.5,0.79]])
```

### cancer

```python
import pandas as pd
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
c_data=load_breast_cancer()
```

```python
c_data.keys()
```

```python
df=pd.DataFrame(c_data['data'])
df.columns=c_data['feature_names']
df['Target']=c_data['target']
df.head(10)
```

```python
df['Target'].value_counts()
```

```python
x2=df[['mean area','mean perimeter','mean concavity','mean symmetry']]
y2=df[['Target']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x2_train,x2_test,y2_train,y2_test=train_test_split(x2,y2,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg2=LogisticRegression()
logreg2.fit(x2_train,y2_train)
```

```python
logreg2.score(x2_train,y2_train)
```

```python
logreg2.score(x2_test,y2_test)
```

```python
logreg2.predict([[1326.0,132.90,0.08690,0.1812]])
```

```python
logreg2.predict_proba([[1326.0,132.90,0.08690,0.1812]])
```
