```python
from sklearn.model_selection import train_test_split
```

```python
knnModel=KNeighborsClassifier(n_neighbors=1)
knnModel.fit(x1_train,y_train)
print("Training Score")
print('\t',knnModel.score(x1_train,y_train)*100)
print("Testing Score")
print('\t',knnModel.score(x1_test,y_test)*100)
```

```python
scores={}
for neighbor in range(1,21):
    knnModel=KNeighborsClassifier(n_neighbors=neighbor)
    knnModel.fit(x1_train,y_train)
    scores[neighbor]=knnModel.score(x1_test,y_test)*100
scores
    
```
